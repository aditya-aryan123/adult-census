```python
from itertools import product
import numpy as np
import scipy.stats as ss
```

```python
cat_var1 = tuple(cat_cols)
cat_var2 = tuple(cat_cols)
```

```python
cat_var_prod = list(product(cat_var1,cat_var2, repeat = 1))
cat_var_prod
```

```python
result = []
for i in cat_var_prod:
    if i[0] != i[1]:
        result.append((i[0], i[1], list(ss.chi2_contingency(
                                        pd.crosstab(cat_df[i[0]], cat_df[i[1]])))[1]))
result
```

```python
chi_test_output = pd.DataFrame(result, columns = ['var1', 'var2', 'coeff'])
chi_test_output.pivot(index='var1', columns='var2', values='coeff')
```

```python
from sklearn.cluster import KMeans
```

```python
copy_df = df.copy()
```

```python
cat_cols = [col for col in df.columns if copy_df[col].dtypes == 'object']
label_encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    le.fit(copy_df[col])
    copy_df[col] = le.transform(copy_df[col])
    label_encoders[col] = le
```

```python
scores = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit_predict(copy_df)
    scores.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(range(1, 11), scores)
plt.show()
```

```python
kl = KneeLocator(
    range(1, 11), scores, curve="convex", direction="decreasing"
)
```

```python
kl.elbow
```
